# 🧠 Single Agent Pipeline Project

## Problem Statement
Build a **Single-Agent Smart Assistant** that:
- Understands user queries
- Routes tasks based on intent
- Uses tools when required
- Returns structured JSON output

### The agent should handle:
- Math queries → Calculator Tool
- Keyword extraction → Keyword Tool
- General queries → Direct response

---
### 🛠️ What You Need to Implement
- Agent logic
- Conditional routing
- Tool integration
- Basic error handling

### 🚀 Bonus
- Improve routing
- Add logging

## 🤖 Implement Agent Logic Below

👉 Use conditional routing:
- If query contains "calculate" → use calculator
- If query contains "keywords" → use keyword extractor
- Else → general response

## 📦 Expected Output Format

```
{
  "type": "calculation / keywords / general / error",
  "result": ...
}
```

````python
import re
import json
````

````python
def calculator_tool(expression):
    try:
        expression = expression.replace("^", "**")
        result = eval(expression)
        return result
    except Exception:
        return "Invalid mathematical expression."
````

````python
def keyword_tool(text):

    stop_words = {
        "the", "is", "a", "an", "of", "to", "and", "in",
        "on", "for", "with", "what", "who", "are", "was",
        "were", "how", "can", "you", "please", "tell", "me"
    }

    words = re.findall(r"\b[a-zA-Z]+\b", text.lower())

    keywords = [
        word for word in words
        if word not in stop_words
    ]

    return list(dict.fromkeys(keywords))
````

````python
def general_response(query):
    return f"I understand your question: '{query}'. This is a general response."
````

````python
def detect_intent(query):

    query = query.lower()

    math_pattern = r'^[\d\s\+\-\*\/\(\)\.^]+$'

    if re.match(math_pattern, query.strip()):
        return "math"

    elif any(word in query for word in
             ["calculate", "solve", "add", "subtract", "multiply", "divide"]):
        return "math"

    elif any(word in query for word in
             ["keyword", "keywords", "extract"]):
        return "keyword"

    else:
        return "general"
````

````python
def smart_agent(query):

    try:

        intent = detect_intent(query)

        if intent == "math":

            expression = query.lower()
            expression = expression.replace("calculate", "")
            expression = expression.replace("solve", "")

            result = calculator_tool(expression)

        elif intent == "keyword":

            result = keyword_tool(query)

        else:

            result = general_response(query)

        output = {
            "query": query,
            "intent": intent,
            "result": result,
            "status": "success"
        }

    except Exception as e:

        output = {
            "query": query,
            "intent": "unknown",
            "result": None,
            "status": "error",
            "message": str(e)
        }

    return output
````

````python
query = "calculate 25 * 8 + 12"

response = smart_agent(query)

print(json.dumps(response, indent=4))
````

````python
query = "Extract keywords from Artificial Intelligence is transforming healthcare"

response = smart_agent(query)

print(json.dumps(response, indent=4))
````

````python
query = "What is Machine Learning?"

response = smart_agent(query)

print(json.dumps(response, indent=4))
````

````python
while True:

    user_query = input("Enter your query (type 'exit' to stop): ")

    if user_query.lower() == "exit":
        print("Goodbye!")
        break

    response = smart_agent(user_query)

    print(json.dumps(response, indent=4))
````
